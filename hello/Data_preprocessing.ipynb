```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
datai = iris.data
feature_names = iris.feature_names

dfi = pd.DataFrame(datai, columns = feature_names)
```

```python
def min_max_normalize(dfi):
    imin = np.min(dfi, axis=0)
    imax = np.max(dfi,axis=0)
    return(dfi - imin)/(imax-imin)
```

```python
mscaled = min_max_normalize(dfi)
print("normalized iris data (min-max scaled:\n",mscaled)
print("\n")
```

```python
kf = KFold(n_splits=20, shuffle=False,random_state = None)

for fold, (train_idx,test_idx) in enumerate(kf.split(mscaled)):
    print(f"K-Fold {fold+1}")

    train_data,test_data = dfi.iloc[train_idx],dfi.iloc[test_idx]

    train_minmax = min_max_normalize(train_data)
    test_minmax = min_max_normalize(test_data)

    print("train")
    print("Train Data (Min-Max Scaled):\n", train_minmax)
    print("\n")
    print("Test Data (Min-Max Scaled):\n", test_minmax)
    print("\n")
    
```
